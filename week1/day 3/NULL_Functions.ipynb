```sql
SELECT * 
FROM Emps
WHERE salary IS NULL;
```

```sql
SELECT * 
FROM Orderss
WHERE discount IS NOT NULL;
```

```sql
SELECT * 
FROM Products
WHERE category IS NULL;
```

```sql
SELECT COUNT(*) 
FROM Emps
WHERE manager_id IS NULL;
```

```sql
SELECT emp_id, IFNULL(salary, 0) AS salary
FROM Emps;
```

```sql
SELECT emp_id, IFNULL(bonus, 1000) AS bonus
FROM Emps;
```

```sql
SELECT order_id, IFNULL(amount, 500) AS amount
FROM Orderss;
```

```sql
SELECT product_id, IFNULL(stock, 0) AS stock
FROM Products;
```

```sql
SELECT name, COALESCE(salary, bonus) AS earnings
FROM Emps;
```

```sql
SELECT name, COALESCE(salary, bonus, 0) AS income
FROM Emps;
```

```sql
SELECT product_name, COALESCE(price, 1000) AS price
FROM Products;
```

```sql
SELECT customer_name, COALESCE(amount, discount, 0) AS payment
FROM Orderss;
```

```sql
SELECT NULLIF(salary, 0)
FROM Emps;
```

```sql
SELECT NULLIF(discount, 0)
FROM Orderss;
```

```sql
SELECT amount / NULLIF(discount, 0)
FROM Orderss;
```

```sql
SELECT NULLIF(coupon_code, 'DISC10')
FROM Orderss;
```

```sql
SELECT name,
COALESCE(salary, 0) + COALESCE(bonus, 0) AS total_earnings
FROM Emps;
```
