```sql
SELECT 
    emp_id,
    SUM(amount) AS total_sales
FROM Sales
GROUP BY emp_id;
```

```sql
SELECT 
    emp_id,
    COUNT(*) AS sales_count
FROM Sales
GROUP BY emp_id;
```

```sql
SELECT 
    product,
    SUM(amount) AS total_sales
FROM Sales
GROUP BY product;
```

```sql
SELECT 
    product,
    AVG(amount) AS avg_sales
FROM Sales
GROUP BY product;
```

```sql
SELECT 
    emp_name,
    COUNT(*) AS sales_count
FROM Employee
JOIN Sales ON Employee.emp_id = Sales.emp_id 
GROUP BY emp_name
HAVING COUNT(*) > 2;
```

```sql
SELECT 
    e.emp_id,
    SUM(e.salary) AS total_salary,
    SUM(s.amount) AS total_sales
FROM Employee e
JOIN Sales s 
ON e.emp_id = s.emp_id
GROUP BY e.emp_id;
```

```sql
SELECT 
    emp_id,
    COUNT(DISTINCT product) AS unique_products
FROM Sales
GROUP BY emp_id;
```

```sql
SELECT 
    emp_id,
    MAX(amount) AS max_sale
FROM Sales
GROUP BY emp_id;
```

```sql
SELECT 
    product,
    SUM(amount) AS total_sales
FROM Sales
GROUP BY product
HAVING SUM(amount) > 50000;
```

```sql
SELECT 
    e.department,
    AVG(s.amount) AS avg_sales
FROM Employee e
JOIN Sales s 
ON e.emp_id = s.emp_id
GROUP BY e.department
ORDER BY avg_sales DESC
LIMIT 1;
```

```sql
SELECT 
    e.department,
    SUM(s.amount) AS total_sales
FROM Employee e
JOIN Sales s 
ON e.emp_id = s.emp_id
GROUP BY e.department
ORDER BY total_sales DESC
LIMIT 1;
```

```sql
SELECT 
    e.emp_name ,
    SUM(s.amount) AS total_sales
FROM Employee e
JOIN Sales s 
ON e.emp_id = s.emp_id
GROUP BY e.emp_name 
ORDER BY total_sales DESC
LIMIT 3;
```

```sql
SELECT 
    YEAR(joining_date) AS join_year,
    COUNT(*) AS total_employees,
    AVG(salary) AS avg_salary
FROM Employee
GROUP BY YEAR(joining_date);
```

```sql
SELECT 
    e.department,
    SUM(s.amount) AS total_sales
FROM Employee e
JOIN Sales s 
ON e.emp_id = s.emp_id
GROUP BY e.department;
```

```sql
SELECT 
    e.department,
    e.emp_id
FROM Employee e
LEFT JOIN Sales s 
ON e.emp_id = s.emp_id
WHERE s.emp_id IS NULL;
```
