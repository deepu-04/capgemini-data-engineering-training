```sql
SELECT e.emp_name, SUM(s.amount) AS total_sales
FROM Employee e
JOIN Sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_name;
```

```sql
SELECT e.emp_name, COUNT(s.sales_id) AS total_sales_count
FROM Employee e
JOIN Sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_name;
```

```sql
SELECT product, SUM(amount) AS total_sales
FROM Sales
GROUP BY product;
```

```sql
SELECT e.emp_name, COUNT(s.sales_id) AS total_sales
FROM Employee e
JOIN Sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_name
HAVING COUNT(s.sales_id) > 2;
```

```sql
SELECT e.emp_id, e.emp_name,
       e.salary,
       COALESCE(SUM(s.amount), 0) AS total_sales
FROM Employee e
LEFT JOIN Sales s
ON e.emp_id = s.emp_id
GROUP BY e.emp_id, e.emp_name, e.salary;
```

```sql
SELECT e.emp_name,
       COUNT(DISTINCT s.product) AS unique_products
FROM Employee e
JOIN Sales s
ON e.emp_id = s.emp_id
GROUP BY e.emp_name;
```

```sql
SELECT e.emp_name,
       MAX(s.amount) AS highest_sale
FROM Employee e
JOIN Sales s
ON e.emp_id = s.emp_id
GROUP BY e.emp_name;
```

```sql
SELECT product,
       SUM(amount) AS total_sales
FROM Sales
GROUP BY product
HAVING SUM(amount) > 50000;
```

```sql
SELECT e.department,
       AVG(s.amount) AS avg_sales
FROM Employee e
JOIN Sales s
ON e.emp_id = s.emp_id
GROUP BY e.department
ORDER BY avg_sales DESC;
```

```sql
SELECT e.department,
       SUM(s.amount) AS total_sales
FROM Employee e
JOIN Sales s
ON e.emp_id = s.emp_id
GROUP BY e.department
ORDER BY total_sales DESC
LIMIT 1;
```

```sql
SELECT e.emp_name,
       SUM(s.amount) AS total_sales
FROM Employee e
JOIN Sales s
ON e.emp_id = s.emp_id
GROUP BY e.emp_name
ORDER BY total_sales DESC
LIMIT 3;
```

```sql
SELECT YEAR(joining_date) AS joining_year,
       COUNT(*) AS total_employees,
       AVG(salary) AS avg_salary
FROM Employee
GROUP BY YEAR(joining_date);
```

```sql
SELECT e.department,
       SUM(s.amount) AS total_sales
FROM Employee e
JOIN Sales s
ON e.emp_id = s.emp_id
GROUP BY e.department;
```

```sql
SELECT e.department,
       COUNT(e.emp_id) AS employees_without_sales
FROM Employee e
LEFT JOIN Sales s
ON e.emp_id = s.emp_id
WHERE s.emp_id IS NULL
GROUP BY e.department;
```
