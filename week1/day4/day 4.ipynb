```python
from pyspark.sql.functions import col, lower, trim

# Clean column names
responses_clean = task1_responses_raw.toDF(
    *[c.strip() for c in task1_responses_raw.columns]
)

# Normalize responses
responses_df = responses_clean.select(
    lower(trim(col("Email address"))).alias("email"),
    col("Timestamp")
)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

window_spec = Window.partitionBy("student_id").orderBy(col("Timestamp"))

duplicates_df = responses_with_id.withColumn(
    "row_num",
    row_number().over(window_spec)
)
```

```python
duplicate_records = duplicates_df.filter(col("row_num") > 1)

display(duplicate_records)
```

```python
first_submission = responses_with_id.groupBy("student_id") \
    .agg({"Timestamp": "min"}) \
    .withColumnRenamed("min(Timestamp)", "first_timestamp")
```

```python
grouped_result = responses_with_id.join(
    first_submission,
    on="student_id",
    how="left"
)
```

```python
valid_responses = responses_df.filter(
    col("Phase 1 Folder Link").isNotNull() &
    (trim(col("Phase 1 Folder Link")) != "") &
    (~lower(col("Phase 1 Folder Link")).isin("na", "null", "n/a"))
)
```

```python
valid_with_id = valid_responses.join(
    email_mapping,
    on="email",
    how="left"
)
```

```python
window_spec = Window.partitionBy("student_id").orderBy(col("Timestamp"))

classified_df = valid_with_id.withColumn(
    "row_num",
    row_number().over(window_spec)
)
```

```python
classified_df = classified_df.withColumn(
    "status",
    when(col("student_id").isNull(), "Invalid")  # email not in master
    .when(col("row_num") == 1, "Submitted")     # first valid submission
    .otherwise("Duplicate")                     # extra submissions
)
```

```python
submitted_students = classified_df.select("student_id").filter(col("student_id").isNotNull()).distinct()

not_submitted = master_df.select("student_id").join(
    submitted_students,
    on="student_id",
    how="left_anti"
).withColumn("status", col("student_id") * 0)  # temp
from pyspark.sql.functions import lit

not_submitted = not_submitted.withColumn("status", lit("Not Submitted"))
```

```python
final_result = classified_df.select("student_id", "email", "status") \
    .unionByName(not_submitted.select("student_id", col("student_id").cast("string").alias("email"), "status"))

display(final_result)
```
